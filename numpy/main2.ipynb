```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# data
#y=2*x -1
years_of_experiences=np.array([0,1,2,3,4,6,8,10]).reshape(-1,1)
salaries=np.array([30000,40000,55000,60000,70000,80000,85000,87000])
```

```python
# model
from sklearn.tree import DecisionTreeRegressor
model=DecisionTreeRegressor(max_depth=2)   #هرچی عمق بیشتر بشه دقت میره بالاتر 
model.fit(years_of_experiences,salaries)
y_predict=model.predict(years_of_experiences)

plt.plot(years_of_experiences, salaries, "*")
plt.plot(years_of_experiences, y_predict)
plt.show()
```
