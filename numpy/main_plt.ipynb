```python
import matplotlib.pyplot as plt
```

```python
#  نمودار خطی 

x=[1,2,3,4,5,6,7,8,9]
y=[2,3,5,7,9,9,7,6,5]
plt.plot(x,y, "o" ,color="red")
plt.plot(x,y, "*" ,color="red")
plt.plot(x,y,color="red")
plt.plot(x,y, marker="o" ,color="red")
plt.title("eexample")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
x=["red", "green", "blue", "gray"]
y=[1,3,7,8]
plt.bar(x,y)

plt.plot(x,y,color="red")
plt.title("example")
plt.xlabel("color")
plt.ylabel("count")
plt.show()
```
