```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('./diabetes.csv')
df.head()
```

```python
df.shape
```

```python
df['Outcome'].value_counts
```

```python
x=df.drop('Outcome',axis=1)
y=df['Outcome']
```

x=np.array(x)
y=np.array(y)

#normalize


```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x=scaler.fit_transform(x)
```

#train / test

```python
from sklearn.model_selection import train_test_split
x_train,x_test , y_train,  y_test=train_test_split(x,y,test_size=0.2)
```

```python
x_train.shape,y_train.shape
```

```python
x_test.shape, y_test.shape
```

```python
# Naive bayes
```

```python
from sklearn.naive_bayes import GaussianNB
model=GaussianNB()
model.fit(x_train , y_train)
```

```python
y_pred_train=model.predict(x_train)
y_pred_test=model.predict
```

```python
from sklearn.metrics import accuracy_score
acc_train=accuracy_score(y_train, y_pred=y_pred_train)
acc_test=accuracy_score(y_test, y_pred=y_pred_test)
acc_train,acc_test
```

```python
from sklearn.metrics import confusion_matrix , precision_score ,recall_score
confusion_matrix(y_test ,y_pred_test)
```

```python
p=precision_score(y_test , y_pred_test)
p
```

```python
r=recall_score(y_test , y_pred_test)
r
```

```python
#knn
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn=KNeighborsClassifier(n_neighbors=23)   #  بهترین حالته که recall درصد بیشتری داره .
knn.fit(x_train , y_train)
```

```python
y_pred_train=knn.predict(x_train)
```

```python
y_pred_train=knn.predict(x_train)
y_pred_test=knn.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
acc_train=accuracy_score(y_train , y_pred_train)
acc_test=accuracy_score(y_test , y_pred_test)

acc_train ,acc_test

# دقت تو داده های ترین خیلی بالاتر و بهتر از تست هستش 
```
