```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import pylab as pl 
%matplotlib inline
```

```python
df=pd.read_csv("FuelConsumption.csv")
# take a look at the dataset
df.head()
# df.head(10)
```

```python
cdf=df[['ENGINESIZE', 'CYLINDERS', 'FUELCONSUMPTION_COMB', 'CO2EMISSIONS', 'FUELCONSUMPTION_CITY', 'FUELCONSUMPTION_HWY' ]]
cdf.head(9)
```

```python
plt.scatter(cdf.ENGINESIZE, cdf.CO2EMISSIONS, color='red')
plt.xlabel('Engine size')
plt.ylabel('Emission')
plt.show()
```

```python
msk=np.random.rand(len(df)) <0.8
train=cdf[msk]
trest=cdf[~msk]
```

# multiple regression

```python
from sklearn import linear_model
regr=linear_model.LinearRegression()
x=np.asanyarray(train[['ENGINESIZE', 'CYLINDERS', 'FUELCONSUMPTION_COMB']])   # من این ایکس ها رو دارم بر حسب این ایکس ها y های پایین رو برام پیدا کن 
y=np.asanyarray(train[['CO2EMISSIONS']])   #بر حسب ایکس های بالا این  yها رو برام پیدا کن 
regr.fit(x,y)          # لطفا فیت کن ایکس هام رو با y
# the coeficient
print('Coefficients:', regr.coef_)
print('Intercept :', regr.intercept_)      # formula => 64 +11*x1 + 7 *x2  + 9*x3 
```
